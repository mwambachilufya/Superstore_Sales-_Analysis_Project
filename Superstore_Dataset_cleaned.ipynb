```python
duplicates = df.duplicated(subset=['order_id', 'product_id', 'customer_id'], keep=False)
print(df[duplicates])
```

```python
aggregated_columns = ['sales', 'quantity', 'discount', 'profit']

id_columns = [col for col in df.columns if col not in aggregated_columns and col != 'row_id']

df_clean = df.groupby(id_columns, as_index=False).sum()

print(f"\nOriginal rows: {len(df)}")
print(f"After aggregation: {len(df_clean)}")
print(f"Rows combined: {len(df) - len(df_clean)}")
```

```python
print(df_clean[df_clean['order_id'] == 'CA-2013-140571'])
```

```python
df.describe(include ='all')
```

```python
for col in ['sales', 'quantity', 'discount', 'profit']:
    Q1, Q3 = df_clean[col].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    outliers = df_clean[(df_clean[col] < Q1 - 1.5*IQR) | (df_clean[col] > Q3 + 1.5*IQR)]
    print(f"{col}: {len(outliers)} outliers")
```

```python
df_clean['order_date'] = pd.to_datetime(df_clean['order_date'])
df_clean['year'] = df_clean['order_date'].dt.year
df_clean['month'] = df_clean['order_date'].dt.month

df_clean['profit_margin'] = (df_clean['profit'] / df_clean['sales']) * 100

df_clean['has_discount'] = df_clean['discount'] > 0
```

```python
df_clean.to_csv('Superstore_Dataset_cleaned.csv', index=False)
```
